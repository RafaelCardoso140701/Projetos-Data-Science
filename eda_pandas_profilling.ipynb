 ## Pandas Proliling

*texto em itálico*
- **Pandas Profiling**: Um dos mais utilizados, gera relatórios detalhados com estatísticas descritivas, análise de correlação, contagem de valores, matriz de valores ausentes e visualizações.

- **O que ele faz?**

Em vez de você ter que olhar coluna por coluna e fazer cálculos manuais, o Pandas Profiling faz tudo sozinho.

 Irá mostrar:

- Quantas linhas e colunas tem o seu conjunto de dados

- Quantos valores faltando (nulos) existem

- Quais colunas têm valores repetidos

- Gráficos com a distribuição dos dados (ex: histogramas)

- Relações entre colunas (ex: correlação entre variáveis)

- Tipos de dados (números, textos, datas etc.)
